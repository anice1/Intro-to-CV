# Intro to Computer Vision 

### (Grayscale Spectrum)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.arange(255)
image
```

```python
image = image[np.newaxis,:]
image.shape
```

```python
image = np.repeat(image, 255, axis=0)
image
```

```python
# plt.figure(figsize=(100, 7))
plt.imshow(image,cmap='gray')
```
